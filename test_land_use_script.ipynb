```python
import scripts.read_land_use as land_use
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
array, unique_classes = land_use.read_land_use()
```

```python
plt.figure(figsize=(12, 12))
plt.axis('off')
im = plt.imshow(array, cmap=plt.cm.get_cmap('Set2', len(unique_classes)))
plt.colorbar(ticks=[0,1,2,3,4,5,6], format=plt.FuncFormatter(lambda val, loc: unique_classes[val]))
plt.show()
```
